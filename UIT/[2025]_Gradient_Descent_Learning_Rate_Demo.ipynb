$x^* = \underset{x\in\mathbb{R}}{\arg\min}~3x^2+4x-2$

```python
import matplotlib.pyplot as plt # library for plotting
import numpy as np # library for vectors, matrices, linear algebra
```

```python
# define a function
def f1(x):
    return 3*x**2 + 4*x - 2
```

```python
x = np.linspace(-3, 3, 100) # Create a array of 100 equally-spaced numbers from -3 to 3
plt.figure()
plt.plot(x, f1(x))
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
```

```python
def gradient_f1(x):
    return 6.0 * x + 4.0
```

```python
def gradient_descent(f, gradient, x_init, step_size, max_iters=50):
    all = [] # all steps of the algorithm running
    x = x_init
    iter = 0
    while True:
        value = f(x)
        grad = gradient(x)
        print(f'iter:{iter}     x={x:.5f};      f(x)={value:.5f};       f\'(x)={grad:5f}')
        all.append((x, value, grad))  # Saving the result
        x = x - step_size * grad
        iter += 1
        if iter > max_iters:
            break
        if np.fabs(grad) < 1e-5:
            break

    return all
```

Let's try different learning rates (step sizes) for Gradient Descent.

```python
all = gradient_descent(f1, gradient_f1, x_init=-2, step_size=1.0/3.0, max_iters=100)
```

```python
all = gradient_descent(f1, gradient_f1, x_init=-2, step_size=1.0/3.0-0.001, max_iters=1000)
```

```python
all = gradient_descent(f1, gradient_f1, x_init=-2, step_size=1.0/10.0, max_iters=100)
```

```python
all = gradient_descent(f1, gradient_f1, x_init=-2, step_size=1.0/6.0, max_iters=100)
```

```python
import time
from IPython import display
```

```python
x = np.linspace(-3, 3, 100) # Create a array of 100 equally-spaced numbers from -3 to 3

for iteration in range(0, len(all)):
    plt.figure()
    ax = plt.gca()
    ax.set_aspect(0.1) # ratio of height to width ?
    plt.plot(x, f1(x))
    plt.xlabel('x', fontsize=15)
    plt.ylabel('f(x)', fontsize=15)
    plt.xlim(-3,3)
    plt.ylim(-0.5,9)

    plt.scatter(all[iteration][0], all[iteration][1], c='red', s=50)
    plt.show()
    time.sleep(1)
    if iteration < len(all):
        display.clear_output(wait=True)
```

$x^*=\underset{x\in\mathbb{R}}{\arg\min}~x^4+2.5x^3-0.5x^2-x+1$

```python
def f2(x):
    return x**4 + 2.5*x**3 - 0.5*x**2 - x + 1.0
```

```python
x = np.linspace(-3, 3, 100) # Create a array of 100 equally-spaced numbers from -3 to 3
plt.figure()
ax = plt.gca()
ax.set_aspect(0.5) # ratio of height to width ?
plt.plot(x, f2(x))
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
plt.xlim(-4,2)
plt.ylim(-4,4)
```

```python
def gradient_f2(x):
    return 4.0*x**3 + 3.0*2.5*x**2 - x - 1.0
```

```python
def gradient_descent(f, gradient, x_init, step_size, max_iters=50):
    all = [] # all steps of the algorithm running
    x = x_init
    iter = 0
    while True:
        value = f(x)
        grad = gradient(x)
        print(f'iter:{iter}     x={x:.5f};      f(x)={value:.5f};       f\'(x)={grad:5f}')
        all.append((x, value, grad))  # Saving the result
        x = x - step_size * grad
        iter += 1
        if iter > max_iters:
            break
        if np.fabs(grad) < 1e-5:
            break

    return all
```

```python
all = gradient_descent(f2, gradient_f2, x_init=-0.5, step_size=0.03, max_iters=50)
```

```python
import time
from IPython import display
```

```python
x = np.linspace(-3, 3, 100) # Create a array of 100 equally-spaced numbers from -3 to 3

for iteration in range(0, len(all)):
    plt.figure()
    ax = plt.gca()
    ax.set_aspect(0.5) # ratio of height to width ?
    plt.plot(x, f2(x))
    plt.xlabel('x', fontsize=15)
    plt.ylabel('f(x)', fontsize=15)
    plt.xlim(-4,2)
    plt.ylim(-4,4)

    plt.scatter(all[iteration][0], all[iteration][1], c='red', s=50)
    plt.show()
    time.sleep(0.1)
    if iteration < len(all):
        display.clear_output(wait=True)
```

```python
def Ackley(x,y):
    return -20.0 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + np.e + 20
```

```python
x = np.linspace(-3,3, 100)
y = np.linspace(-3,3, 100)
X, Y= np.meshgrid(x,y)
Z = Ackley(X, Y)
plt.figure(figsize=(12,6))
ax1 = plt.subplot(1,2,1, projection='3d')
surf = ax1.plot_surface(X,Y,Z, cmap='jet', alpha=0.7)
ax2 = plt.subplot(1,2,2)
contour = ax2.contourf(X,Y,Z, levels=50, cmap='jet', alpha=0.7)
```
